```python
import numpy as np
from fdfdpy.Fdfd import Fdfd

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

## Waveguide example

```python
omega = 2*np.pi*200e12
dl = 0.01 # grid size (units of L0, which defaults to 1e-6)
eps_r = np.ones((201,201)) # relative permittivity
eps_r[:,95:105] = 12.25 # set waveguide region
NPML = [15, 15] # number of pml grid points on x and y borders

simulation = Fdfd(omega, eps_r, dl, NPML, 'Ez')
simulation.add_mode(3.5, 'x', [16, 100], 40)
simulation.setup_modes()
simulation.solve_fields()

simulation.plt_re(outline=True, cbar=True);
```
